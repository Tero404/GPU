```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal



device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

Schr√∂dinger eq. $$-\frac{\hbar^2}{2m} \frac{d^2\psi}{dx^2} + V(x)\psi=E\psi$$ $$\psi(0)=\psi(L)=0 $$

for python $\hbar = 1$ and the formula is multipied by $m$. So we get:
$$-\frac{1}{2}\frac{d^2\psi}{dx^2}+mV(x)\psi=mE\psi$$

```python
N = 2000
dy = 1/N
y = np.linspace(0,1,N+1)
```

```python
def mL2V(y):
    return 1000*(y-0.5)**2
```

```python
d = 1/dy**2 + mL2V(y)[1:-1]
e = -1/(2*dy**2) * np.ones(len(d)-1)
```

```python
w,v = eigh_tridiagonal(d,e)
```

```python
plt.plot(v.T[0]**2)
plt.plot(v.T[1]**2)
plt.plot(v.T[2]**2)
plt.plot(v.T[3]**2)
#
```
